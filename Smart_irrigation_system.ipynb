// Pin Definitions
int soilSensorPin = A0;   // Soil moisture sensor connected to analog pin A0
int ledPin = 7;           // Green LED connected to digital pin 7
int motorPin = 8;         // Motor control pin through transistor
int buzzerPin = 6;        // Buzzer connected to digital pin 6
int threshold = 500;      // Moisture threshold value (adjust if needed)

void setup() {
  pinMode(ledPin, OUTPUT);    // Set LED pin as output
  pinMode(motorPin, OUTPUT);  // Set motor pin as output
  pinMode(buzzerPin, OUTPUT); // Set buzzer pin as output
  Serial.begin(9600);         // Start serial communication for debugging
}

void loop() {
  int moisture = analogRead(soilSensorPin);  // Read soil moisture value
  Serial.print("Soil Moisture: ");
  Serial.println(moisture);

  if (moisture < threshold) {
    // Soil is dry
    digitalWrite(ledPin, HIGH);    // Turn LED ON
    digitalWrite(motorPin, HIGH);  // Turn motor ON via transistor
    tone(buzzerPin, 1000);         // Turn buzzer ON (1kHz tone)
    Serial.println("Soil Dry - Watering ON ðŸ’§ðŸ”Š");
  } else {
    // Soil is wet
    digitalWrite(ledPin, LOW);     // Turn LED OFF
    digitalWrite(motorPin, LOW);   // Turn motor OFF
    noTone(buzzerPin);             // Turn buzzer OFF
    Serial.println("Soil Wet - Watering OFF âœ…");
  }

  delay(1000);  // Wait 1 second before next reading
}
