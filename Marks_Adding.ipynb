```python
# Create the input.csv file with the provided data
data = """Roll Num,Marks
102003542,1
102003640,3
102003725,1
102003307,1
102003643,1
102017134,1
"""

with open("input.csv", "w") as f:
    f.write(data)

print("input.csv created succesfully.")
```

```python
import csv
from collections import defaultdict

# Step 1: Read input from a file (input.csv)
marks_dict = defaultdict(int)

with open("input.csv", mode="r") as file:
    reader = csv.reader(file)
    next(reader)  # Skip header
    for row in reader:
        roll_num = row[0]
        marks = int(row[1])
        marks_dict[roll_num] += marks

# Step 2: Write to output file (output.csv)
with open("output.csv", mode="w", newline='') as file:
    writer = csv.writer(file)
    writer.writerow(["Roll Num", "Marks"])
    for roll_num, total_marks in marks_dict.items():
        writer.writerow([roll_num, total_marks])

print("âœ… Marks have been summed and written to output.csv")
```
