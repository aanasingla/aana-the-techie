float tempC;

void setup()
{
  pinMode(A0, INPUT);       // Temperature sensor input
  pinMode(11, OUTPUT);      // Buzzer or Fan
  pinMode(13, OUTPUT);      // LED or Heater
  Serial.begin(9600);       // Start serial monitor
}

void loop()
{
  int tempRaw = analogRead(A0);              // Read analog value
  float voltage = (tempRaw * 5.0) / 1023.0;   // Convert to voltage
  tempC = voltage * 100;                      // LM35: 10mV per °C

  Serial.print("Temperature: ");
  Serial.print(tempC);
  Serial.println(" °C");

  if (tempC > 25)
  {
    tone(11, 500);           // Buzzer ON
    digitalWrite(13, HIGH);  // Turn on LED/Heater
  }
  else
  {
    noTone(11);              // Buzzer OFF
    digitalWrite(13, LOW);   // Turn off LED/Heater
  }

  delay(1000);  // Wait 1 second
}
