#include <LiquidCrystal.h>

// Initialize LCD (RS, E, D4, D5, D6, D7)
LiquidCrystal lcd(7, 6, 5, 4, 3, 2);

int sensorEntrance = 8; // IR Sensor 1 OUT
int sensorExit = 9;     // IR Sensor 2 OUT

int peopleCount = 0;

void setup() {
  pinMode(sensorEntrance, INPUT);
  pinMode(sensorExit, INPUT);

  lcd.begin(16, 2);
  lcd.print("Visitor Counter");
  delay(2000);
  lcd.clear();
  lcd.print("People Inside:");
  lcd.setCursor(0, 1);
  lcd.print(peopleCount);

  Serial.begin(9600);
}

void loop() {
  if (digitalRead(sensorEntrance) == LOW) {
    delay(300); // Debounce
    if (digitalRead(sensorExit) == LOW) {
      peopleCount++;
      updateLCD();
      Serial.print("Entered: ");
      Serial.println(peopleCount);
      delay(1000);
    }
  }

  if (digitalRead(sensorExit) == LOW) {
    delay(300); // Debounce
    if (digitalRead(sensorEntrance) == LOW) {
      if (peopleCount > 0) {
        peopleCount--;
        updateLCD();
        Serial.print("Exited: ");
        Serial.println(peopleCount);
        delay(1000);
      }
    }
  }
}

void updateLCD() {
  lcd.clear();
  lcd.print("People Inside:");
  lcd.setCursor(0, 1);
  lcd.print(peopleCount);
}
